```python
#지역별 전문의 증감량
agg_dict = {
    '내과' : 'sum',
    '신경과' : 'sum',
    '정신건강의학과' : 'sum',
    '외과' : 'sum',
    '정형외과' : 'sum',
    '신경외과' : 'sum',
    '심장혈관흉부외과' : 'sum',
    '성형외과' : 'sum',
    '마취통증의학과' : 'sum',
    '산부인과' : 'sum',
    '소아청소년과' : 'sum',
    '안과' : 'sum',
    '이비인후과' : 'sum',
    '피부과' : 'sum',
    '비뇨의학과' : 'sum',
    '영상의학과' : 'sum',
    '방사선종양학과' : 'sum',
    '진단검사의학과' : 'sum',
    '재활의학과' : 'sum',
    '핵의학과' : 'sum',
    '가정의학과' : 'sum',
    '응급의학과' : 'sum',
    '직업환경의학과' : 'sum',
    '예방의학과' : 'sum'
}


tmp = df.groupby(['시도', '요양기관종별', '연도']).agg(agg_dict).reset_index()
mask = tmp['요양기관종별'].isin(['계', '상급종합병원', '종합병원', '병원', '의원']) #유의미한 병원 유형만 옮겼습니다

tmp = tmp.loc[mask, :]

# 2021년에 4개마다 나오는지 확인
for i in range(0, len(tmp['연도'].tolist()), 4):
    if tmp['연도'].tolist()[i] != 2021:
        print(i)
        break
    if i == len(tmp['연도'].tolist())-4: 
        print('이상무')
        test = tmp.copy()

tmp_lst = tmp['연도'].tolist()
n = len(tmp_lst)

for i in range(0, n, 4): # 4개 행마다 끊어서 pct_change() 적용
    test.iloc[i:i+4, 3:] = (test.iloc[i:i+4, 3:].pct_change() * 100).round(2)

test = pd.concat([test.iloc[:,:3], test.iloc[:, 3:].add_suffix('_증감률')], axis=1) # 내과~ 부터 suffix
# test.head(15)
test
```

```python
# 집계 딕셔너리 설정 (기존과 동일)
agg_dict = {
    '내과' : 'sum',
    '신경과' : 'sum',
    '정신건강의학과' : 'sum',
    '외과' : 'sum',
    '정형외과' : 'sum',
    '신경외과' : 'sum',
    '심장혈관흉부외과' : 'sum',
    '성형외과' : 'sum',
    '마취통증의학과' : 'sum',
    '산부인과' : 'sum',
    '소아청소년과' : 'sum',
    '안과' : 'sum',
    '이비인후과' : 'sum',
    '피부과' : 'sum',
    '비뇨의학과' : 'sum',
    '영상의학과' : 'sum',
    '방사선종양학과' : 'sum',
    '진단검사의학과' : 'sum',
    '재활의학과' : 'sum',
    '핵의학과' : 'sum',
    '가정의학과' : 'sum',
    '응급의학과' : 'sum',
    '직업환경의학과' : 'sum',
    '예방의학과' : 'sum'
}

# 연도별 합계를 지역별 및 요양기관종별로 그룹화
specialty_by_region = df.groupby(['시도', '요양기관종별', '연도']).agg(agg_dict).reset_index()

# 2021년과 2024년 데이터를 각각 필터링
data_2021 = specialty_by_region[specialty_by_region['연도'] == 2021]
data_2024 = specialty_by_region[specialty_by_region['연도'] == 2024]

# 2021년과 2024년 데이터를 병합하여 비교
comparison = pd.merge(data_2021, data_2024, on=['시도', '요양기관종별'], suffixes=('_2021', '_2024'))

# 각 과목별 증감률 계산
for specialty in agg_dict.keys():
    comparison[f'{specialty}_증감'] = (
        (comparison[f'{specialty}_2024'] - comparison[f'{specialty}_2021']) / comparison[f'{specialty}_2021'] * 100
    ).round(2)

# 결과 확인
comparison
```
