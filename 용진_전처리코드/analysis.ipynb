```python
df_simple = df.drop(['진료과목코드명', '과목별 전문의수'], axis=1).drop_duplicates()
df_simple.shape[0] == df_simple['암호화요양기호'].nunique()
```

```python
df_simple['총병상수'] = df_simple.iloc[:, 7:21].sum(axis=1)

agg_dict = {
    '총병상수': 'sum',
    '총의사수': 'sum',
    '암호화요양기호' : 'count'
}


dnb_table = df_simple.groupby(['시도코드명','시군구코드명']).agg(agg_dict).reset_index()
dnb_table.rename(columns={'시도코드명':'행정구역', '암호화요양기호':'총병원수'}, inplace=True)

dnb_table = pd.merge(dnb_table, pop_df.iloc[1:, :2], how='left', on='행정구역')
dnb_table.rename(columns={'2024년09월_계_총인구수':'총인구수'}, inplace=True)

dnb_table['총인구수'] = [ int(''.join(dnb_table['총인구수'][i].split(','))) for i in range(dnb_table.shape[0]) ]
dnb_table.info()
```

```python
dnb_table
```

```python
agg_dict = {
    '총병상수' : 'sum',
    '총의사수' : 'sum',
    '총병원수' : 'sum',
    '총인구수' : 'mean'
}

tmp = dnb_table.groupby('행정구역').agg(agg_dict).reset_index()

# 병상수 / 인구수
tmp['인구대비병상수'] = tmp['총병상수'] / tmp['총인구수']

# 의사수 / 인구수
tmp['인구대비의사수'] = tmp['총의사수'] / tmp['총인구수']

# 병원수 / 인구수
tmp['인구대비병원수'] = tmp['총병원수'] / tmp['총인구수']
```
