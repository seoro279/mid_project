```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/hong/Desktop/practice_code/mid_project/data/전처리 자료/시군구별 인구&병원.csv')
df.head()
```

```python
df['총인구수'] = df['총인구수'].str.replace(',', '')
df['총인구수'] = df['총인구수'].astype(int)
```

```python
tmp = df.groupby(['시도코드명','시군구코드명'])['총인구수'].mean().reset_index()
```

```python
tmp
```

```python
pop = tmp.groupby('시도코드명')['총인구수'].sum().reset_index()
```

```python
mask_class = df['종별코드명'] == '상급종합'
mask_area = df['시도코드명'].isin(['서울', '경기', '인천'])

tophost_cap = df.loc[(mask_class) & (mask_area), :]
tophost_nocap = df.loc[(mask_class) & (~mask_area), :]
```

```python
tophost_nocap.head()
```

```python
mask_area2 = pop['시도코드명'].isin(['서울', '경기', '인천'])
```

```python
print('비수도권 10만명당 상급종합병원 병상수')
print(tophost_nocap['총병상수'].sum() / pop.loc[~mask_area2, '총인구수'].sum().sum() * 100000)
print('-'*80)
print('비수도권 10만명당 상급종합병원 병상수')
print(tophost_cap['총병상수'].sum() / pop.loc[mask_area2, '총인구수'].sum().sum() * 100000)
# 차이가 거의 없음,,
```

```python
print('비수도권 10만명당 상급종합병원 병상수')
print(tophost_cap['총병상수'].sum() / pop.loc[mask_area2, '총인구수'].sum().sum() * 100000)
```
