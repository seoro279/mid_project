```python
agg_dict = {
    '내과' : 'sum',
    '신경과' : 'sum',
    '정신건강의학과' : 'sum',
    '외과' : 'sum',
    '정형외과' : 'sum',
    '신경외과' : 'sum',
    '심장혈관흉부외과' : 'sum',
    '성형외과' : 'sum',
    '마취통증의학과' : 'sum',
    '산부인과' : 'sum',
    '소아청소년과' : 'sum',
    '안과' : 'sum',
    '이비인후과' : 'sum',
    '피부과' : 'sum',
    '비뇨의학과' : 'sum',
    '영상의학과' : 'sum',
    '방사선종양학과' : 'sum',
    '진단검사의학과' : 'sum',
    '재활의학과' : 'sum',
    '핵의학과' : 'sum',
    '가정의학과' : 'sum',
    '응급의학과' : 'sum',
    '직업환경의학과' : 'sum',
    '예방의학과' : 'sum'
}


tmp = df.groupby(['시도', '요양기관종별', '연도']).agg(agg_dict).reset_index()
mask = tmp['요양기관종별'].isin(['상급종합병원', '종합병원', '병원', '의원']) #유의미한 병원 유형만 옮겼습니다

tmp = tmp.loc[mask, :]

# 2021년에 4개마다 나오는지 확인
for i in range(0, len(tmp['연도'].tolist()), 4):
    if tmp['연도'].tolist()[i] != 2021:
        print(i)
        break
    if i == len(tmp['연도'].tolist())-4: 
        print('이상무')
        test = tmp.copy()

tmp_lst = tmp['연도'].tolist()
n = len(tmp_lst)

for i in range(0, n, 4):
    test.iloc[i:i+4, 3:] = (test.iloc[i:i+4, 3:].pct_change() * 100).round(2)

test = pd.concat([test.iloc[:,:3], test.iloc[:, 3:].add_suffix('_증감률')], axis=1)
test.head(15)
```
