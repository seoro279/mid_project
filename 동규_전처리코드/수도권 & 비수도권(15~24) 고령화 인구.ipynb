```python
import pandas as pd
```

```python
df_2024_oldage = pd.read_excel('202409_고령인구현황_OldAge.xlsx')
df_2024_oldage
```

```python
len(df_2024_oldage.iloc[1:,:]['행정구역'].sort_values())
```

```python
# 202409_고령인구현황_OldAge 와 지역별_인구대비_병상의사병원수의 행정구역명 차이 발생 > 통일시키기
df_2024_oldage['행정구역'] = ['전국', '서울', '부산', '대구', '인천', '광주', '대전', '울산', '세종', '경기', '강원', '충북', '충남', '전북', '전남', '경북', '경남', '제주']
# df_oldage['행정구역']
df_2024_oldage

# 고령인구와 지역별 간 행의 개수 차이 > 고령인구 내 '전국' 제외 > 행정구역명 순서 통일시키기
df_not_whole = df_2024_oldage.iloc[1:,:]
df_not_whole.sort_values('행정구역')
```

```python
# 행정구역별 65세이상 그룹화
col_65_or_more = df_not_whole.groupby('행정구역')['2024년09월_65세이상전체'].sum().reset_index()
col_65_or_more
```

```python
# 지역별_인구대비_병상의사병원수 불러오기
df_beddochospi = pd.read_csv('지역별_인구대비_병상의사병원수.csv')
len(df_beddochospi['행정구역'])
```

```python
# unname 제거
cols = df_beddochospi.iloc[:,1:]
cols
```

```python
# 고령대비병상수, 고령대비총의사수, 고령대비총병원수
col_65_or_more_table = pd.merge(cols, col_65_or_more ,how = 'left', on = '행정구역')
col_65_or_more_table['고령대비병상수'] = col_65_or_more_table['총병상수'] / col_65_or_more_table['2024년09월_65세이상전체']
col_65_or_more_table['고령대비총의사수'] = col_65_or_more_table['총의사수'] / col_65_or_more_table['2024년09월_65세이상전체']
col_65_or_more_table['고령대비총병원수'] = col_65_or_more_table['총병원수'] / col_65_or_more_table['2024년09월_65세이상전체']
col_65_or_more_table['지역별고령비율'] = col_65_or_more_table['총인구수'] / col_65_or_more_table['2024년09월_65세이상전체']
col_65_or_more_table
```

```python
col_65_or_more_table.to_csv('65_or_more.csv', index = False)
```

```python
per_year_oldage['2024년_계_총인구수'] = df_2024_oldage['2024년09월_전체']
per_year_oldage['2023년_계_총인구수'] = df_2023_oldage['2023년_계_총인구수']
per_year_oldage['2022년_계_총인구수'] = df_2022_oldage['2022년_계_총인구수']
per_year_oldage['2021년_계_총인구수'] = df_2021_oldage['2021년_계_총인구수']
per_year_oldage['2020년_계_총인구수'] = df_2020_oldage['2020년_계_총인구수']
per_year_oldage['2019년_계_총인구수'] = df_2019_oldage['2019년_계_총인구수']
per_year_oldage['2018년_계_총인구수'] = df_2018_oldage['2018년_계_총인구수']
per_year_oldage['2017년_계_총인구수'] = df_2017_oldage['2017년_계_총인구수']
per_year_oldage['2016년_계_총인구수'] = df_2016_oldage['2016년_계_총인구수']
per_year_oldage['2015년_계_총인구수'] = df_2015_oldage['2015년_계_총인구수']
per_year_oldage
```

```python
per_year_oldage.columns
```

```python
# 수도권 고령화
mask1= per_year_oldage['행정구역'].isin(['서울', '경기', '인천'])
mask1

capital_region = per_year_oldage.loc[mask1,:].reset_index()
capital_region = capital_region.iloc[:,1:]
capital_region
```

```python
capital_region.to_csv('수도권고령화.csv', index = False)
```

```python
# 비수도권 고령화
mask2 = ~per_year_oldage['행정구역'].isin(['서울', '경기', '인천'])
mask2
noncapital_region = per_year_oldage.loc[mask2,:].reset_index()
noncapital_region = noncapital_region.iloc[:,1:]
noncapital_region
```

```python
noncapital_region.to_csv('비수도권고령화.csv', index = False)
```

```python
# 수도권 & 비수도권(15~24) 고령화 인구
capi_gap = pd.concat([capital_region, noncapital_region], axis = 0, ignore_index=True)
capi_gap
```

```python
capi_gap.to_csv('수도권 & 비수도권(15~24) 고령화 인구.csv', index = False)
```
